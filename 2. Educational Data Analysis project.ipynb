# Data Filtering

```python
# Filter the DataFrame and save it to 'df_remove'
df_remove = df[df['Income_Group'] == 'Upper middle income']

# Preview the first few rows of 'df_remove'
df_remove.head()
```

```python
df_remove["Income_Group"].value_counts() #Counts 'Income_Group' in df_remove.
```

```python
df_remove =df_remove.reset_index(drop=True) #Resets 'df_remove' index.
df_remove.head() #Displays the first 5 rows of df_recet.
```

```python
column_to_check = '2-alpha_code'# Replace 'ema' with the correct column name if it's different

# Generate random 2-letter code only where 'column_to_check' is null
df_remove['2-alpha_code'] = df_remove['2-alpha_code'].where(
    df_remove[column_to_check].notnull(), 
    [''.join(random.choices(string.ascii_uppercase, k=2)) for _ in range(df_remove[column_to_check].isnull().sum())]
)
# Display the updated DataFrame
df_remove.head()
```

```python
column_to_check = 'WB-2_code'# Replace 'ema' with the correct column name if it's different

# Generate random 2-letter code only where 'column_to_check' is null
df_remove['WB-2_code'] = df_remove['WB-2_code'].where(
    df_remove[column_to_check].notnull(), 
    [''.join(random.choices(string.ascii_uppercase, k=2)) for _ in range(df_remove[column_to_check].isnull().sum())]
)
# Display the updated DataFrame
df_remove.head()
```
